```python
import turtle
import math
t = turtle.Turtle()
nha_cao = 100
nha_rong = 80
nha_dai = 60
thancay_ngan = 10
thancay_dai = 20
tanla_canhngan = 30
mainha_canhngan = math.sqrt(nha_rong*nha_rong/2)
tanla_canhdai = math.sqrt(tanla_canhngan*tanla_canhngan*2)
duongkinh_mattroi = 20
# t.speed(50)

#-------Vẽ ngôi nhà ---------
t.penup()
t.goto(50,50)
t.pendown()
#Vẽ mặt trước ngôi nhà
t.fillcolor('blue')
t.begin_fill()
t.forward(nha_rong)
t.left(90)
t.forward(nha_cao)
t.left(90)
t.forward(nha_rong)
t.left(90)
t.forward(nha_cao)
t.end_fill()
#Vẽ cửa ngôi nhà
t.fillcolor('green')
t.begin_fill()
t.left(90)
t.forward(nha_rong/3)
t.left(90)
t.forward(nha_rong/2)
t.right(90)
t.forward(nha_rong/3)
t.right(90)
t.forward(nha_rong/2)
t.left(90)
t.forward(nha_rong/3)
t.end_fill()
#Vẽ mặt bên phải ngôi nhà
t.fillcolor('yellow')
t.begin_fill()
t.left(30)
t.forward(nha_dai)
t.left(60)
t.forward(nha_cao)
t.left(120)
t.forward(nha_dai)
t.end_fill()
#Vẽ mái nhà
t.fillcolor('pink')
t.begin_fill()
t.right(75)
t.forward(mainha_canhngan)
t.left(90)
t.forward(mainha_canhngan)
t.end_fill()
t.left(180)
t.forward(mainha_canhngan)
t.right(15)
t.fillcolor('orange')
t.begin_fill()
t.forward(nha_dai)
t.right(75)
t.forward(mainha_canhngan)
t.right(105)
t.forward(nha_dai)
t.right(75)
t.forward(mainha_canhngan)
t.penup()
t.right(105)
t.forward(mainha_canhngan)
t.right(75)
t.forward(nha_dai)
t.pendown()
t.end_fill()
#Vẽ cửa sổ ngôi nhà
t.fillcolor('brown')
t.begin_fill()
t.right(65)
t.penup()
t.forward(nha_dai/2)
t.pendown()
t.right(40)
t.forward(nha_dai/2)
t.left(60)
t.forward(nha_dai/2)
t.left(120)
t.forward(nha_dai/2)
t.left(60)
t.forward(nha_dai/2)
t.end_fill()

# ------ Vẽ ông mặt trời --------
t.penup()
t.forward(100)
t.right(90)
t.forward(20)
t.pendown()
t.forward(duongkinh_mattroi*2)
t.backward(duongkinh_mattroi)
t.left(90)
t.forward(duongkinh_mattroi)
t.backward(duongkinh_mattroi*2)
t.forward(duongkinh_mattroi)
t.left(45)
t.forward(duongkinh_mattroi/2+6)
t.backward(duongkinh_mattroi+12)
t.forward(duongkinh_mattroi/2+6)
t.left(90)
t.forward(duongkinh_mattroi/2+6)
t.backward(duongkinh_mattroi+12)
t.forward(6)
t.right(90)
t.fillcolor('yellow')
t.begin_fill()
t.circle(duongkinh_mattroi/2)
t.end_fill()

#-------Vẽ cây thông------
t.penup()
t.left(90+45)
t.forward(200)
t.pendown()
#Vẽ thân cây
t.left(90)
t.fillcolor('brown')
t.begin_fill()
t.forward(thancay_ngan)
t.left(90)
t.forward(thancay_dai)
t.left(90)
t.forward(thancay_ngan)
t.left(90)
t.forward(thancay_dai)
t.end_fill()
#Vẽ lá cây
t.penup()
t.right(180)
t.forward(math.sqrt(tanla_canhngan*tanla_canhngan-(tanla_canhdai/2)*(tanla_canhdai/2))*3 + thancay_dai)
t.right(90)
t.forward(thancay_ngan/2)
t.right(45)
t.pendown()
t.fillcolor('green')
t.begin_fill()
#Vẽ tán lá đầu tiên
t.forward(tanla_canhngan)
t.right(90+45)
t.forward(tanla_canhdai)
t.right(90+45)
t.forward(tanla_canhngan)
t.right(90+45)
t.penup()
t.forward(math.sqrt(tanla_canhngan*tanla_canhngan-(tanla_canhdai/2)*(tanla_canhdai/2)))
#Vẽ tán lá thứ 2
t.pendown()
t.left(45)
t.forward(tanla_canhngan)
t.right(90+45)
t.forward(tanla_canhdai)
t.right(90+45)
t.forward(tanla_canhngan)
t.right(90+45)
t.penup()
t.forward(math.sqrt(tanla_canhngan*tanla_canhngan-(tanla_canhdai/2)*(tanla_canhdai/2)))
#Vẽ tán lá thứ 3
t.pendown()
t.left(45)
t.forward(tanla_canhngan)
t.right(90+45)
t.forward(tanla_canhdai)
t.right(90+45)
t.forward(tanla_canhngan)
t.right(90+45)
t.penup()
t.forward(math.sqrt(tanla_canhngan*tanla_canhdai/2))
t.end_fill()
turtle.done()
```
